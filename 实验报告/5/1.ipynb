```python
# 计算相关系数
def corrcoef(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    cov = np.sum((x-x_mean)*(y-y_mean))
    return cov/(np.sqrt(np.sum((x-x_mean)**2))*np.sqrt(np.sum((y-y_mean)**2)))  
corrcoef(arr0, arr1)
```
